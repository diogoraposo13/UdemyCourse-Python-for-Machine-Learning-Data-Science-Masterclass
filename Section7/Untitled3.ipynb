```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(10,8), dpi=200)
sns.boxplot(data=df,x='reading score', y='parental level of education', hue='test preparation course', palette='Set2')

plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
plt.figure(figsize=(10,12), dpi=200)
sns.violinplot(data=df,x='reading score', y='parental level of education',hue='test preparation course', palette='Set2', split=True, inner='quartile',bw=0.3)

plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
plt.figure(figsize=(8,4), dpi=200)
sns.swarmplot(data=df,x='math score', y='gender', size=2, hue='test preparation course')

plt.legend(bbox_to_anchor=(1.25,0.5))
```
