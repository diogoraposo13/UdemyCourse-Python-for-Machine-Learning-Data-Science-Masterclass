```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b=a**4
```

```python
x = np.arange(0,10)
y=2*x
```

```python
fig, axes = plt.subplots(2,2, figsize=(4,1), dpi=200)

axes[0][0].plot(a,b)

axes[0][1].plot(a,b)

axes[1][0].plot(a,b)
axes[1][0].set_xlabel('X label')
axes[1][0].set_ylabel('Y label')
axes[1][0].set_title('Title')
axes[1][0].set_xlim(0,15)
axes[1][0].set_ylim(0,15000)

axes[1][1].plot(a,b)

fig.suptitle("Figure Level Title")

#plt.tight_layout()

#fig.subplots_adjust(wspace=1, hspace=1)


#plt.tight_layout()
fig.set_figwidth(10)

fig.savefig('new_subplots.png', bbox_inches='tight')

```
