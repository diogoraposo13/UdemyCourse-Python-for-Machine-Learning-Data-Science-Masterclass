```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/cluster_mpg.csv')
```

```python
df_w_dummies = pd.get_dummies(df.drop('name',axis=1))
```

```python
df_w_dummies
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#MinMaxScaler

# sqrt((x1_a - x1_b)^2 )
```

```python
scaler = MinMaxScaler()
```

```python
scaled_data = scaler.fit_transform(df_w_dummies)
```

```python
scaled_data
```

```python
scaled_df = pd.DataFrame(scaled_data,columns=df_w_dummies.columns)
```

```python
scaled_df
```

```python
plt.figure(figsize=(15,8))
sns.clustermap(scaled_df, col_cluster=False)
```

```python
len(df)
```

```python
from sklearn.cluster import AgglomerativeClusteringrativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=3)
```

```python
#scaled_df
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(data=df,x='mpg',y='horsepower', hue=cluster_labels, palette='viridis')
```

```python
model = AgglomerativeClustering(n_clusters=None, distance_threshold=0)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

```python
linkage_matrix = hierarchy.linkage(model.children_)
```

```python
linkage_matrix # 2 clusters joining together | distance between the 2 clusters | number of points falling in this particular linkage 
```

```python
plt.figure(figsize=(20,10))
dendro = dendrogram(linkage_matrix, truncate_mode='lastp',p=11)
```
